## Sinusfunktion
$y=sin(x)$

Cosinus: $y=cos(x)$

Varianten: $y = a (sin(bx+c)) + d$

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sbn
%matplotlib inline    
mpl.rcParams['figure.figsize'] = (16,6)
#plt.figure(figsize=(16, 6))

x = np.arange(0,4*np.pi,0.01)   # start,stop,step
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
y2 = np.cos(x)
plt.plot(x,y)
plt.plot(x,y2)
plt.show()
```

```python
y3 = np.sin(x +np.pi)
y4 = np.sin(2*x)
y5 = 0.25 * np.sin(8*x)

plt.plot(x,y)
plt.plot(x,y3)
plt.plot(x,y4)
plt.plot(x,y5)
plt.show()
```

## Überlagerung mehrerer Sinuskurven

$y = sin(sin(x * a) * b)$

```python
y6 = np.sin(np.sin(np.sin(np.sin(x) * 2) * 1.6)* 1.3)
y7 = np.sin(np.sin(np.sin(x) * np.pi) * np.pi)

plt.plot(x,y)
plt.plot(x,y6)
plt.plot(x,y7)
plt.show()
```

## Darstellung als Fourier-Reihe

$
\begin{align}
f(x)=\sum_{n=1,3,5,7...}^\infty \Big(\frac{4}{n\pi}\Big)sin\Big(\frac{n \pi x}{L}\Big)
\end{align}
$

```python
def f1(x,n):
    return (4 / n * np.pi) * np.sin(n * np.pi * x / 5)
    

def f2(x,n):
    ns = np.arange(1,n,2)
    return np.sum([f1(x, nn) for nn in ns])
    
yn = lambda n: np.array([f2(xi, n) for xi in x])

ys = np.array([yn(n) for n in np.arange(3,10,2)])
ys = np.swapaxes(ys,0,1)
print(ys)

#plt.plot(x,y)
plt.plot(x,ys)
plt.show()
```
